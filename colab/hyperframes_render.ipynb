## Step 3 — Run the Hyperframes render

**PROJECT** = folder name inside repo (e.g. `pikachu`)
**OUTPUT** = output filename

````
PROJECT = 'pikachu'   # ← change this to your project folder
OUTPUT  = 'pikachu_final.mp4'

import subprocess, os

result = subprocess.run(
    ['hyperframes', 'render', '--workers', '1', '--fps', '30',
     '--quality', 'high', PROJECT, '-o', OUTPUT],
    capture_output=True, text=True,
    cwd='/content/hyperframes_workspace'
)
print(result.stdout)
if result.returncode != 0:
    print('STDERR:', result.stderr)
````

## Step 4 — Download the rendered video to your machine

````
from google.colab import files

output_path = f'/content/hyperframes_workspace/{OUTPUT}'
print(f'File size: {os.path.getsize(output_path) / 1024 / 1024:.1f} MB')
files.download(output_path)
print('Download started — check your browser downloads!')
````
